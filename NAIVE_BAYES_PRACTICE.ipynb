```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('spam.csv',names=['class','messages'],encoding='latin1')
```

```python
data.head()
```

```python
data['length']=data['messages'].apply(len)
data['length']
```

```python
data.head()
```

```python
data.describe()
```

```python
data[data['length']==910]['messages'].iloc[0]
```

```python
data[data['length']==2]['messages'].iloc[0]
```

```python
data['target']=data['class'].map({'spam':0,'ham':1})
```

```python
data.head()
```

```python
data.loc[data['target'].isnull(),'target']=0.0
```

```python
data.head()
```

```python
data.drop(0,axis=0,inplace=True)
```

```python
data.head()
```

```python
data.reset_index(drop=True,inplace=True)
```

```python
data.head()
```

```python
import string
```

```python
def remove_punct(text):
    text="".join([char for char in text if char not in string.punctuation])
    return text.lower()
```

```python
data['clean_text']=data['messages'].apply(lambda x:remove_punct(x))
```

```python
data.head()
```

```python
data['clean_text']
```

```python
data['messages']
```

```python
x=data['clean_text']
y=data['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
cv=CountVectorizer()
```

```python
x_cv=cv.fit_transform(x_train)
```

```python
y_cv=cv.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb=MultinomialNB()
```

```python
nb.fit(x_cv,y_train)
```

```python
predict=nb.predict(y_cv)
```

```python
predict
```

```python
from sklearn.metrics import classification_report
```

```python
report=classification_report(predict,y_test)
print(report)
```

```python
message=input('enter message:')
input_msg=cv.transform([message])
predict=nb.predict(input_msg)
if (predict[0]==0):
    print('Fake Message')
else:
    print('True Message')
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
tf=TfidfVectorizer()
```

```python
x_tf=tf.fit_transform(x_train)
```

```python
y_tf=tf.transform(x_test)
```

```python
be=BernoulliNB()
```

```python
be.fit(x_tf,y_train)
```

```python
predict1=be.predict(y_tf)
```

```python
predict1
```

```python
report1=classification_report(predict1,y_test)
```

```python
print(report1)
```

```python
message=input('Enter the Message')
input_=tf.transform([message])
predict=be.predict(input_)
if (predict[0]==0):
    print('THIS MESSAGE IS FAKE')
else:
    print('THIS IS TRUE MESSAGE')
```
