```python
from textblob.classifiers import NaiveBayesClassifier
from textblob import TextBlob
from sklearn.model_selection  import train_test_split
```

```python
reviews = {
    "Inception": [
        {"review": "Amazing visuals and story!", "sentiment": "positive"},
        {"review": "A bit confusing but great movie.", "sentiment": "positive"},
        {"review": "Too hard to follow, not enjoyable.", "sentiment": "negative"},
    ],
    "Titanic": [
        {"review": "A beautiful and emotional film.", "sentiment": "positive"},
        {"review": "Too long and boring.", "sentiment": "negative"},
        {"review": "Heartbreaking and powerful.", "sentiment": "positive"},
    ],
    "The Dark Knight": [
        {"review": "Heath Ledger was legendary!", "sentiment": "positive"},
        {"review": "The best superhero movie ever.", "sentiment": "positive"},
        {"review": "Too dark and slow.", "sentiment": "negative"},
    ]
}
```

```python
texts = []
labels = []

for movie, review_list in reviews.items():
    for r in review_list:
        texts.append(r["review"])
        labels.append(r["sentiment"])
```

```python
x_train,x_test,y_train,y_test=train_test_split(texts,labels,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
train_data=list(zip(x_train,y_train))
test_data=list(zip(x_test,y_test))
```

```python
from textblob import download_corpora
download_corpora.download_all()

```

```python
classifier=NaiveBayesClassifier(train_data)
```

```python
classifier
```

```python
text=input('Enter reviews')
blob=TextBlob(text,classifier=classifier)
print(blob.classify())
print(blob.sentiment)
```
