```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
brain=sns.load_dataset('brain_networks').iloc[3:,1:]
brain.head()
```

```python
brain.reset_index(drop=True,inplace=True)
```

```python
brain.head()
```

```python
brain=brain.astype(float)
brain.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_brain=scaler.fit_transform(brain)
```

```python
brain=pd.DataFrame(data=scaled_brain,columns=brain.columns)
```

```python
brain.head()
```

```python
from sklearn.decomposition import PCA
pca=PCA()
principle_component=pca.fit_transform(brain)
plt.figure
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('no of components')
plt.ylabel('varience')
plt.title('explained_varience_ratio')
plt.show()
```

```python
np.where(np.cumsum(pca.explained_variance_ratio_)>0.95)
```

```python
pca1=PCA(n_components=43)
new_brain=pca1.fit_transform(brain)
n_component=43
new_col=[f'principal component{i}' for i in range(1,n_component+1) ]
brain_new=pd.DataFrame(data=new_brain,columns=new_col)
```
